```python
# 定义函数
def calculate_mean(data):
    return sum(data) / len(data)  # sum()和len()是Python内置的

# 调用函数：传入压力那一列
mean_pressure = calculate_mean(df["pressure"])
print(f"平均压力是: {mean_pressure}")
```

```python
# --- Step 4: 写异常判断函数 ---
def detect_anomaly(value):
    if value > 200:
        return True
    else:
        return False

# --- Step 5: 找出异常点 ---
anomalies = [] # 就像 C 语言里的动态数组

# 遍历 pressure 这一列，p 就是每一个具体的压力值
for p in df["pressure"]:
    if detect_anomaly(p):
        anomalies.append(p) # 如果是异常，就“压入”列表

print("检测到的异常点有：", anomalies)
```

```python
# --- Step 6: 标记异常位置 (索引) ---
anomaly_index = []
for i in range(len(df)):
    if detect_anomaly(df["pressure"][i]):
        anomaly_index.append(i)

# --- Step 7: 画图并标红 ---
import matplotlib.pyplot as plt

plt.plot(df["time"], df["pressure"], label="Pressure Trend")

# 用 C 语言的循环思维，在特定的索引位置打点
for i in anomaly_index:
    plt.scatter(df["time"][i], df["pressure"][i], color="red", zorder=5)

plt.xlabel("Time")
plt.ylabel("Pressure")
plt.title("Anomaly Detection Result")
plt.legend()
plt.show()
```
